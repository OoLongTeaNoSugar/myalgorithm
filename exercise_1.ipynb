```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 读取数据

```python
data = pd.read_csv("data/height.vs.temperature.csv")
data.head()
```

```python
data.columns
```

# 数据可视化

```python
plt.figure(figsize=(16, 8))
plt.scatter(data[u'height'], data[u'temperature'], c ='black')
plt.xlabel("Height")
plt.ylabel("Temperature")
plt.show()
```

# 训练模型

```python
X = data[u'height'].values.reshape(-1,1)
y = data[u'temperature'].values.reshape(-1,1)

reg = LinearRegression()
reg.fit(X, y)
```

```python
print('a = {:.5}'.format(reg.coef_[0][0]))
print('b = {:.5}'.format(reg.intercept_[0]))

print("线性模型为: Y = {:.5}X + {:.5} ".format(reg.coef_[0][0], reg.intercept_[0]))
```

```python
predictions = reg.predict(X)

plt.figure(figsize=(16, 8))
plt.scatter(data[u'height'], data[u'temperature'], c ='black')
plt.plot(data[u'height'], predictions,c ='blue', linewidth=2)
plt.xlabel("Height")
plt.ylabel("Temperature")
plt.show()
```

```python
# 预测
predictions = reg.predict([[10000]])
print('一万米高空的温度为{:.2}度'.format( predictions[0][0]) )
```
